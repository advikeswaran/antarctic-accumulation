```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.lines as mlines
import numpy as np

data = np.array(pd.read_csv('sealevel.csv'))
seaUncertai = np.array(pd.read_csv('newUncertain.csv'))

apz1 = np.polyfit(data[:70, 0], data[:70, 1], 1)
app1 = np.poly1d(apz1)

apz2 = np.polyfit(data[69:, 0], data[69:, 1], 1)
app2 = np.poly1d(apz2)

waz1 = np.polyfit(data[:70, 0], data[:70, 2], 1)
wap1 = np.poly1d(waz1)

waz2 = np.polyfit(data[69:, 0], data[69:, 2], 1)
wap2 = np.poly1d(waz2)

eaz1 = np.polyfit(data[:70, 0], data[:70, 3], 1)
eap1 = np.poly1d(eaz1)

eaz2 = np.polyfit(data[69:, 0], data[69:, 3], 1)
eap2 = np.poly1d(eaz2)

aiz1 = np.polyfit(data[:70, 0], data[:70, 4], 1)
aip1 = np.poly1d(aiz1)

aiz2 = np.polyfit(data[69:, 0], data[69:, 4], 1)
aip2 = np.poly1d(aiz2)


#plt.plot(data[:70, 0], app1(data[:70, 0]), color='gold', linewidth=3, linestyle='dotted')
#plt.plot(data[69:, 0], app2(data[69:, 0]), color='gold', linewidth=3, linestyle='dotted')
plt.plot(data[:, 0], data[:, 1], color='gold', linewidth=4)



#plt.plot(data[:70, 0], wap1(data[:70, 0]), color='pink', linewidth=3, linestyle='dotted')
#plt.plot(data[69:, 0], wap2(data[69:, 0]), color='pink', linewidth=3, linestyle='dotted')
plt.plot(data[:, 0], data[:, 2], color='magenta', linewidth=4)


#plt.plot(data[:70, 0], eap1(data[:70, 0]), color='skyblue', linewidth=3, linestyle='dotted')
#plt.plot(data[69:, 0], eap2(data[69:, 0]), color='skyblue', linewidth=3, linestyle='dotted')
plt.plot(data[:, 0], data[:, 3], color='deepskyblue', linewidth=4)


#plt.plot(data[:70, 0], aip1(data[:70, 0]), color='gray', linewidth=3, linestyle='dotted')
#plt.plot(data[69:, 0], aip2(data[69:, 0]), color='gray', linewidth=3, linestyle='dotted')
plt.plot(data[:, 0], data[:, 4], color='black', linewidth=10)


    

blu = mlines.Line2D([], [], color='magenta', label='WAIS',
                          markersize=20, lw=5)
red = mlines.Line2D([], [], color='deepskyblue', label='EAIS',
                          markersize=20, lw=5)
green = mlines.Line2D([], [], color='gold', label='AP',
                          markersize=20, lw=5)
black = mlines.Line2D([], [], color='black', label='AIS',
                          markersize=20, lw=5)
plt.legend(handles=[blu, red, green, black], loc='upper left', fontsize=30)



plt.fill_between(data[:, 0], seaUncertai[:, 0], seaUncertai[:, 1], color='lightgray')
    

plt.grid()
plt.margins(x=0)
plt.rcParams["figure.figsize"] = (20,15)
plt.xlabel(" Year ", fontsize=30)
plt.ylabel("Sea Level Mitigation (mm)", fontsize=30)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)

plt.show()
```
